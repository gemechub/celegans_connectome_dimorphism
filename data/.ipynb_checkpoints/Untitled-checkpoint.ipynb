```python
import numpy as np
import pandas as pd
```

```python
cook_herm = pd.read_csv('cook_herm_neurons_anatomical_info.csv')
witv = pd.read_csv('wit_neurons_anatomical_info.csv')
```

```python
cook_herm
```

```python
witv
```

```python
cell_loc = dict(zip(cook_herm.cell_name, cook_herm.general_location))
cell_hem = dict(zip(cook_herm.cell_name, cook_herm.hemisphere))
```

```python
witv['general_location'] = witv['cell_name'].map(cell_loc)
witv['hemisphere'] = witv['cell_name'].map(cell_hem)
```
